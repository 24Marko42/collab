```python
import pandas as pd
import numpy as np
import scipy.stats as st
from google.colab import files

df = pd.read_csv('water.txt', sep='\t')

#Разделение на Север и Юг
df_north = df[df['location'] == 'North']
df_south = df[df['location'] == 'South']

print(f"\n✅ Данные успешно загружены!")
print(f"Северных городов: {len(df_north)}")
print(f"Южных городов: {len(df_south)}")
```

1) Используя метод .describe() вычислите описательные статистики для
северных и южных городов. Сравните средние значения смертности в
северных и южных городах и значения концентрации кальция в питьевой
воде.

```python
print("Описательные статистики для северных городов (смертность и жесткость воды):")
describe_north = df_north[['mortality', 'hardness']].describe()
print(describe_north)

print("\nОписательные статистики для южных городов (смертность и жесткость воды):")
describe_south = df_south[['mortality', 'hardness']].describe()
print(describe_south)

mean_mortality_north = describe_north.loc['mean', 'mortality']
mean_mortality_south = describe_south.loc['mean', 'mortality']
mean_hardness_north = describe_north.loc['mean', 'hardness']
mean_hardness_south = describe_south.loc['mean', 'hardness']

print("\nСравнение средних значений:")
print(f"Средняя смертность в северных городах: {mean_mortality_north:.2f}")
print(f"Средняя смертность в южных городах: {mean_mortality_south:.2f}")
print(f"Разница в средней смертности (Север - Юг): {mean_mortality_north - mean_mortality_south:.2f}")

print(f"\nСредняя концентрация кальция (жесткость) в северных городах: {mean_hardness_north:.2f}")
print(f"Средняя концентрация кальция (жесткость) в южных городах: {mean_hardness_south:.2f}")
print(f"Разница в средней концентрации кальция (Север - Юг): {mean_hardness_north - mean_hardness_south:.2f}")
```

2) Постройте 95% доверительные интервалы для средней годовой смертности
по всем южным и северным городам. Отличаются ли границы интервалов?

```python
mean_mortality_north = describe_north.loc['mean', 'mortality']
std_mortality_north = describe_north.loc['std', 'mortality']
n_north = describe_north.loc['count', 'mortality']

se_mortality_north = std_mortality_north / np.sqrt(n_north)

mortality_ci_north = st.t.interval(confidence=0.95, df=n_north - 1, loc=mean_mortality_north, scale=se_mortality_north)

print(f"95% доверительный интервал для средней смертности в северных городах: {mortality_ci_north[0]:.2f} - {mortality_ci_north[1]:.2f}")

mean_mortality_south = describe_south.loc['mean', 'mortality']
std_mortality_south = describe_south.loc['std', 'mortality']
n_south = describe_south.loc['count', 'mortality']

se_mortality_south = std_mortality_south / np.sqrt(n_south)

mortality_ci_south = st.t.interval(confidence=0.95, df=n_south - 1, loc=mean_mortality_south, scale=se_mortality_south)

print(f"95% доверительный интервал для средней смертности в южных городах: {mortality_ci_south[0]:.2f} - {mortality_ci_south[1]:.2f}")

print("\nСравнение границ интервалов:")
print(f"Интервал для северных городов: [{mortality_ci_north[0]:.2f}, {mortality_ci_north[1]:.2f}]")
print(f"Интервал для южных городов:   [{mortality_ci_south[0]:.2f}, {mortality_ci_south[1]:.2f}]")

if mortality_ci_north[0] <= mortality_ci_south[1] and mortality_ci_north[1] >= mortality_ci_south[0]:
    print("Интервалы пересекаются.")
else:
    print("Интервалы не пересекаются.")
```

3) Постройте 95% доверительные интервалы для средней концентрации
кальция в питьевой воде для южных и северных городов.

```python
mean_hardness_north = describe_north.loc['mean', 'hardness']
std_hardness_north = describe_north.loc['std', 'hardness']
n_north_hardness = describe_north.loc['count', 'hardness']

se_hardness_north = std_hardness_north / np.sqrt(n_north_hardness)

hardness_ci_north = st.t.interval(confidence=0.95, df=n_north_hardness - 1, loc=mean_hardness_north, scale=se_hardness_north)

print(f"95% доверительный интервал для средней концентрации кальция в северных городах: {hardness_ci_north[0]:.2f} - {hardness_ci_north[1]:.2f}")

mean_hardness_south = describe_south.loc['mean', 'hardness']
std_hardness_south = describe_south.loc['std', 'hardness']
n_south_hardness = describe_south.loc['count', 'hardness']

se_hardness_south = std_hardness_south / np.sqrt(n_south_hardness)

hardness_ci_south = st.t.interval(confidence=0.95, df=n_south_hardness - 1, loc=mean_hardness_south, scale=se_hardness_south)

print(f"95% доверительный интервал для средней концентрации кальция в южных городах: {hardness_ci_south[0]:.2f} - {hardness_ci_south[1]:.2f}")

print("\nСравнение границ интервалов:")
print(f"Интервал для северных городов: [{hardness_ci_north[0]:.2f}, {hardness_ci_north[1]:.2f}]")
print(f"Интервал для южных городов:   [{hardness_ci_south[0]:.2f}, {hardness_ci_south[1]:.2f}]")

if hardness_ci_north[0] <= hardness_ci_south[1] and hardness_ci_north[1] >= hardness_ci_south[0]:
    print("Интервалы пересекаются.")
else:
    print("Интервалы не пересекаются.")
```
